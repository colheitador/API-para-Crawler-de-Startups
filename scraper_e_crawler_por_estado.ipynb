```python
# o website startupbase.com.br conta com 10 filtros dentro da categoria "estados"
# vamos utilizar o recurso "for" para percorrer os filtros, mas vamos fazer um "for" para cada filtro, pois o selenium dá muitos erros se pedirmos
# para usar un só "for" para percorrer todos
```

```python
# filtro #10

url_base = "https://startupbase.com.br/home/startups"
driver.get(url_base)

for index_estado in range(10,11):  ## entra só no filtro 10
    #expandir_estados = driver.find_element(By.XPATH, "./html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[1]/div[1]/app-accordion/app-panel[4]").click()
    expandir_estados = driver.find_element(By.XPATH, "./html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[1]/div[1]/app-accordion/app-panel[2]/div/p").click()
    time.sleep(2.5) ## espera esse tempo pra carregar
    
    path_para_numero_de_startups = "./html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[1]/div[1]/app-accordion/app-panel[2]/div/div/ais-refinement-list/div/ul/li[" + str(index_estado) + "]/label/span[2]"
    # print(path_para_numero_de_startups) # printar caso queira ver o path pro numero de startups de cada categoria
    numero_de_startups_por_estado = driver.find_element(By.XPATH, path_para_numero_de_startups).text
    print("numero de startups do estado em análise: ", numero_de_startups_por_estado)
    
    pedacos_path_estado = "./html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[1]/div[1]/app-accordion/app-panel[2]/div/div/ais-refinement-list/div/ul/li[" + str(index_estado) + "]"
    time.sleep(1.5) ## espera esse tempo pra carregar
    #print(pedacos_path_estado)

           # caso o selenium abra algum filtro por contra própria, esse try vai tentar limpar
    try:
        fechar_estado = driver.find_element(By.XPATH, "/html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[2]/section/div/div[2]/div/div/ais-current-refinements/div/button").click()
    except:
        None

    clicar_estado = driver.find_element(By.XPATH, pedacos_path_estado).click()
    fechar_estados = driver.find_element(By.XPATH, "./html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[1]/div[1]/app-accordion/app-panel[2]/div/p").click()
    
 

    element = driver.find_element(By.TAG_NAME, 'body') ## procura qualquer elemento que apareça na página pra usar como referência

    driver.find_element(By.XPATH, "./html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[1]").click()
    #clica num ponto específico da página pra que a rolagem pra baixo se torne possível

    time.sleep(2.5) ## espera esse tempo pra carregar
    
    rolar_pra_baixo_10x()
 
    
    paths_para_links = driver.find_elements(By.XPATH, "./html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[2]/section/ais-infinite-hits/div/div/a")
    
    lista_de_links = []
    for cada_path in paths_para_links:
        cada_link = cada_path.get_attribute('href')
        #print(cada_link) ## mostrar os links obtidos        
        lista_de_links.append(cada_link) # adicionar os links a uma lista para contar quantos links tem
        numero_de_links_da_categoria = len(lista_de_links) ## contar quantos links tem na lista

        
    while numero_de_links_da_categoria < int(numero_de_startups_por_estado):
        print("+-----------------------------------------------------------------------+")
        print("foram encontrados ", numero_de_links_da_categoria, "/", numero_de_startups_por_estado, " links até agora")
        print("realizando nova rolagem pra encontrar mais links \n ...")
        rolar_pra_baixo_10x()
        del paths_para_links
        paths_para_links = driver.find_elements(By.XPATH, "./html/body/app-root/ng-component/app-layout/div/div/div/div/div/app-layout-column/ng-component/div/ais-instantsearch/div/div/div/div[2]/section/ais-infinite-hits/div/div/a")
        lista_de_links = []
        print("fazendo a contagem de links...")
        for cada_path in paths_para_links:
            cada_link = cada_path.get_attribute('href')
            #print(cada_link) ## mostrar os links obtidos        
            lista_de_links.append(cada_link) # adicionar os links a uma lista para contar quantos links tem
        numero_de_links_da_categoria = len(lista_de_links) ## contar quantos links tem na lista

    else:
        print("foram encontrados ", numero_de_links_da_categoria, "/", numero_de_startups_por_estado, " links")
        for cada_path in paths_para_links:
            cada_link = cada_path.get_attribute('href')
            cursor1.execute(f"INSERT INTO tab_links_por_estado VALUES ('{cada_link}')") ## adicionar os links a uma tabela
        
    print("links adicionados à tabela")
```
